```python
import pandas as pd

# Load Data from Google Colab storage
customers_df = pd.read_csv("/content/Customers.csv")
products_df = pd.read_csv("/content/Products.csv")
transactions_df = pd.read_csv("/content/Transactions.csv")

# Inspect the first few rows of each dataset
print("Customers Data:")
# display(customers_df.head())

print("\nProducts Data:")
# display(products_df.head())

print("\nTransactions Data:")
# display(transactions_df.head())

# display method for getting the visualization of the output.
# aliased name.head(), it gives the first few rows of the dataset to confirm the
   # working of the dataset.
```

```python
# Convert date columns to datetime format
customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'], format='%d-%m-%Y', errors='coerce')
transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'], format='%d-%m-%Y %H:%M', errors='coerce')

# Merge Transactions with Products
transactions_merged_df = transactions_df.merge(products_df, on='ProductID', how='left')

#It Merges Transactions with Customers
full_data_df = transactions_merged_df.merge(customers_df, on='CustomerID', how='left')

# It Displays the first few rows of the merged dataset
display(full_data_df.head())
```

```python
# Aggregate customer spending and behavior features
customer_spending = full_data_df.groupby('CustomerID').agg(
    total_spend=('TotalValue', 'sum'),
    avg_spend=('TotalValue', 'mean'),
    total_transactions=('TransactionID', 'count'),
    last_purchase=('TransactionDate', 'max')
).reset_index()

# Compute recency (days since last purchase)
max_transaction_date = full_data_df['TransactionDate'].max()
customer_spending['recency_days'] = (max_transaction_date - customer_spending['last_purchase']).dt.days

# One-hot encode product categories
category_one_hot = pd.get_dummies(full_data_df[['CustomerID', 'Category']], columns=['Category'])
category_prefs = category_one_hot.groupby('CustomerID').sum().reset_index()

# Merge all features into a final customer profile dataset
customer_profile_df = customer_spending.merge(category_prefs, on='CustomerID', how='left')

# Display processed customer profiles
display(customer_profile_df.head())

#What This Does:
# Calculates total spend, transactions, and recency.
# Encodes product category preferences.
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity

# Normalize numerical features
scaler = MinMaxScaler()
numeric_features = ['total_spend', 'avg_spend', 'total_transactions', 'recency_days']
customer_profile_df[numeric_features] = scaler.fit_transform(customer_profile_df[numeric_features])

# Create feature matrix for similarity calculation
feature_cols = numeric_features + [col for col in customer_profile_df.columns if col.startswith("Category_")]
customer_feature_matrix = customer_profile_df[feature_cols]

# Compute cosine similarity
similarity_matrix = cosine_similarity(customer_feature_matrix)

# Convert to DataFrame
customer_ids = customer_profile_df['CustomerID'].values
similarity_df = pd.DataFrame(similarity_matrix, index=customer_ids, columns=customer_ids)

# Display similarity matrix
display(similarity_df.head())

# Manual Min-Max Scaling function
# def min_max_scaling(df, columns):
#     for col in columns:
#         min_val = df[col].min()
#         max_val = df[col].max()
#         df[col] = (df[col] - min_val) / (max_val - min_val)
#     return df

# # Define numerical features to scale
# numeric_features = ['total_spend', 'avg_spend', 'total_transactions', 'recency_days']

# # Apply manual min-max scaling
# customer_profile_df = min_max_scaling(customer_profile_df, numeric_features)

# # Display scaled data
# display(customer_profile_df.head())

```

```python
# Find the top 3 most similar customers for each of the first 20 customers
top_n = 3
lookalike_results = {}

for customer in customer_ids[:20]:  # First 20 customers
    similar_customers = similarity_df[customer].nlargest(top_n + 1).iloc[1:]  # Exclude self
    lookalike_results[customer] = list(zip(similar_customers.index, similar_customers.values))

# Convert results into DataFrame
lookalike_df = pd.DataFrame.from_dict(lookalike_results, orient='index', columns=['Lookalike1', 'Lookalike2', 'Lookalike3'])
lookalike_df = lookalike_df.applymap(lambda x: f"{x[0]} ({x[1]:.4f})")

# Display the Lookalike results
display(lookalike_df.head())
```

```python
# Save to CSV
lookalike_df.to_csv("/content/Lookalike.csv", index_label="CustomerID")

# Download the file
from google.colab import files
files.download("/content/Lookalike.csv")
```
